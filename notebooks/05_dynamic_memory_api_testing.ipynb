```python
import requests
import pandas as pd
import json
import time
import os

print("Libraries imported successfully.")
```

```python
#Set API Base URL

BASE_URL = "http://127.0.0.1:8000"

print("API Base URL:", BASE_URL)
```

```python
#Check API Health

response = requests.get(f"{BASE_URL}/")

print("Status Code:", response.status_code)
print(json.dumps(response.json(), indent=4))
```

```python
#Test Case 1: Add First Wrong Interaction; This tests whether memory is created for a new student.

#Send First Interaction

student_id = 999001
session_id = 5001
concept_name = "Percent Of"

interaction_1 = {
    "student_id": student_id,
    "session_id": session_id,
    "problem_id": 101,
    "concept_name": concept_name,
    "correct": 0,
    "attempt_count": 2,
    "hint_count": 1,
    "hint_total": 3,
    "response_time_ms": 45000
}

response = requests.post(
    f"{BASE_URL}/memory/update",
    json=interaction_1
)

print("Status Code:", response.status_code)
result_1 = response.json()
print(json.dumps(result_1, indent=4))
```

```python
#Convert First Result to Test Summary

test_1 = {
    "test_case": "First wrong interaction",
    "expected_recent_interaction_count": 1,
    "actual_recent_interaction_count": result_1["short_term_memory"].get("recent_interaction_count"),
    "expected_recent_accuracy": 0.0,
    "actual_recent_accuracy": result_1["short_term_memory"].get("recent_accuracy"),
    "expected_correct_count": 0,
    "actual_correct_count": result_1["concept_based_memory"].get("correct_count"),
    "expected_wrong_count": 1,
    "actual_wrong_count": result_1["concept_based_memory"].get("wrong_count"),
    "status": "PASS" if (
        result_1["short_term_memory"].get("recent_interaction_count") == 1 and
        result_1["short_term_memory"].get("recent_accuracy") == 0.0 and
        result_1["concept_based_memory"].get("correct_count") == 0 and
        result_1["concept_based_memory"].get("wrong_count") == 1
    ) else "FAIL"
}

test_1
```

```python
#Test Case 2: Add Second Correct Interaction; This tests whether memory updates after another interaction.

#Send Second Interaction

interaction_2 = {
    "student_id": student_id,
    "session_id": session_id,
    "problem_id": 102,
    "concept_name": concept_name,
    "correct": 1,
    "attempt_count": 1,
    "hint_count": 0,
    "hint_total": 3,
    "response_time_ms": 30000
}

response = requests.post(
    f"{BASE_URL}/memory/update",
    json=interaction_2
)

print("Status Code:", response.status_code)
result_2 = response.json()
print(json.dumps(result_2, indent=4))
```

```python
#Test Second Result

test_2 = {
    "test_case": "Second correct interaction",
    "expected_recent_interaction_count": 2,
    "actual_recent_interaction_count": result_2["short_term_memory"].get("recent_interaction_count"),
    "expected_recent_accuracy": 0.5,
    "actual_recent_accuracy": result_2["short_term_memory"].get("recent_accuracy"),
    "expected_correct_count": 1,
    "actual_correct_count": result_2["concept_based_memory"].get("correct_count"),
    "expected_wrong_count": 1,
    "actual_wrong_count": result_2["concept_based_memory"].get("wrong_count"),
    "status": "PASS" if (
        result_2["short_term_memory"].get("recent_interaction_count") == 2 and
        result_2["short_term_memory"].get("recent_accuracy") == 0.5 and
        result_2["concept_based_memory"].get("correct_count") == 1 and
        result_2["concept_based_memory"].get("wrong_count") == 1
    ) else "FAIL"
}

test_2
```

```python
#Test Case 3: Add Third Correct Interaction; This tests whether session status changes after performance improves.

#Send Third Interaction

interaction_3 = {
    "student_id": student_id,
    "session_id": session_id,
    "problem_id": 103,
    "concept_name": concept_name,
    "correct": 1,
    "attempt_count": 1,
    "hint_count": 0,
    "hint_total": 3,
    "response_time_ms": 25000
}

response = requests.post(
    f"{BASE_URL}/memory/update",
    json=interaction_3
)

print("Status Code:", response.status_code)
result_3 = response.json()
print(json.dumps(result_3, indent=4))
```

```python
#Test Third Result

test_3 = {
    "test_case": "Third correct interaction",
    "expected_recent_interaction_count": 3,
    "actual_recent_interaction_count": result_3["short_term_memory"].get("recent_interaction_count"),
    "expected_correct_count": 2,
    "actual_correct_count": result_3["concept_based_memory"].get("correct_count"),
    "expected_wrong_count": 1,
    "actual_wrong_count": result_3["concept_based_memory"].get("wrong_count"),
    "expected_session_status": "moderate_progress",
    "actual_session_status": result_3["short_term_memory"].get("session_status"),
    "status": "PASS" if (
        result_3["short_term_memory"].get("recent_interaction_count") == 3 and
        result_3["concept_based_memory"].get("correct_count") == 2 and
        result_3["concept_based_memory"].get("wrong_count") == 1 and
        result_3["short_term_memory"].get("session_status") == "moderate_progress"
    ) else "FAIL"
}

test_3
```

```python
#Test Case 4: Add New Concept for Same Student; This tests whether concept-based memory separates concepts correctly.

#Send New Concept Interaction

interaction_4 = {
    "student_id": student_id,
    "session_id": session_id,
    "problem_id": 201,
    "concept_name": "Circle Graph",
    "correct": 0,
    "attempt_count": 3,
    "hint_count": 2,
    "hint_total": 4,
    "response_time_ms": 60000
}

response = requests.post(
    f"{BASE_URL}/memory/update",
    json=interaction_4
)

print("Status Code:", response.status_code)
result_4 = response.json()
print(json.dumps(result_4, indent=4))
```

```python
#Test New Concept Result

test_4 = {
    "test_case": "New concept interaction",
    "expected_target_concept": "Circle Graph",
    "actual_target_concept": result_4.get("target_concept"),
    "expected_total_concepts": 2,
    "actual_total_concepts": result_4["long_term_memory"].get("total_concepts"),
    "expected_circle_graph_interactions": 1,
    "actual_circle_graph_interactions": result_4["concept_based_memory"].get("total_interactions"),
    "status": "PASS" if (
        result_4.get("target_concept") == "Circle Graph" and
        result_4["long_term_memory"].get("total_concepts") == 2 and
        result_4["concept_based_memory"].get("total_interactions") == 1
    ) else "FAIL"
}

test_4
```

```python
#Test Case 5: Retrieve Updated Memory Context; Now check whether the retrieval endpoint returns the updated student memory.

#Retrieve Memory Context

response = requests.get(f"{BASE_URL}/memory/context/{student_id}")

print("Status Code:", response.status_code)
retrieved_context = response.json()
print(json.dumps(retrieved_context, indent=4))
```

```python
#Test Retrieval

test_5 = {
    "test_case": "Retrieve updated memory context",
    "expected_student_id": student_id,
    "actual_student_id": retrieved_context.get("student_id"),
    "has_short_term_memory": "short_term_memory" in retrieved_context,
    "has_long_term_memory": "long_term_memory" in retrieved_context,
    "has_concept_based_memory": "concept_based_memory" in retrieved_context,
    "status": "PASS" if (
        retrieved_context.get("student_id") == student_id and
        "short_term_memory" in retrieved_context and
        "long_term_memory" in retrieved_context and
        "concept_based_memory" in retrieved_context
    ) else "FAIL"
}

test_5
```

```python
#Step 5: Create Final Test Results Table

#Combine All Test Results

test_results = pd.DataFrame([
    test_1,
    test_2,
    test_3,
    test_4,
    test_5
])

test_results
```

```python
#Count Passed and Failed Tests

summary = test_results["status"].value_counts().reset_index()
summary.columns = ["status", "count"]

summary
```

```python
#Step 6: Save Test Outputs

#Save Results

os.makedirs("../outputs/tables", exist_ok=True)
os.makedirs("../outputs/api_results", exist_ok=True)

test_results.to_csv(
    "../outputs/tables/dynamic_memory_api_test_results.csv",
    index=False
)

summary.to_csv(
    "../outputs/tables/dynamic_memory_api_test_summary.csv",
    index=False
)

with open("../outputs/api_results/final_dynamic_memory_context.json", "w") as f:
    json.dump(retrieved_context, f, indent=4, default=str)

print("Dynamic memory API test outputs saved successfully.")
```
